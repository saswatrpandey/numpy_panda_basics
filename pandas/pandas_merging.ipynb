```python
week1 = pd.read_csv('csv_files/Restaurant - Week 1 Sales.csv')
week2 = pd.read_csv('csv_files/Restaurant - Week 2 Sales.csv')
customers = pd.read_csv('csv_files/Restaurant - Customers.csv')
foods = pd.read_csv('csv_files/Restaurant - Foods.csv')
```

```python
print(f'week 1: {len(week1)}, week 2: {len(week2)}')
```

```python
pd.concat([week1, week2], ignore_index= True)
```

```python
sales = pd.concat([week1, week2], keys=['A', 'B']) #even though the index are repeated another index is added A B.
```

```python
sales.loc[('A', 240)]
```

```python
sales.loc[('B', 240), 'Customer ID']
```

INNER JOIN

```python
week1.merge(week2, how='inner', on='Customer ID')
```

```python
week1.merge(week2, how='inner', on='Customer ID', suffixes=['- Week 1', '- Week 2'])
```

```python
week1.merge(week2, how ='inner', on=['Customer ID', 'Food ID'])
```

```python
week1[week1['Customer ID'] == 578]
```

```python
week2[week2['Customer ID'] == 578]
```

```python
merged = week1.merge(week2, how='outer', 
                     on='Customer ID', 
                     suffixes=['- Week 1', '- Week 2'], 
                     indicator= True)
```

```python
merged['_merge'].value_counts()
```

```python
mask = merged['_merge'].isin(['left_only', 'right_only'])
merged[mask]
```
