```python
import pandas as pd
```

```python
df = pd.read_csv('csv_files/employees.csv')
df.head()
```

```python
df.info()
```

Optimization->

```python
df['Start Date'] = pd.to_datetime(df["Start Date"])
```

```python
df['Last Login Time'] = pd.to_datetime(df['Last Login Time'], format='%I:%M %p')
```

```python
df['Senior Management'] = df['Senior Management'].astype('bool')
```

```python
df['Gender'] = df['Gender'].astype('category')
```

```python
df.info()
```

Filter data

```python
df['Gender'] == 'Male'
```

```python
df[df['Gender']=='Male']
```

```python
df[df['Senior Management']]
```

```python
mask = df['Bonus %'] < 1.5
df[mask]
```

```python
df[(df['Gender'] == 'Male') & (df['Team']=='Marketing')]
df[(df['Gender'] == 'Male') | (df['Team']=='Marketing')]
```

```python
#using isin for simpler filters
df[df['Team'].isin(['Legal', 'Sales', 'Product'])] #also can have series,list,tuples
```

```python
df[df['Team'].isnull()]
df[df['Team'].notnull()]
```

```python
#inbetween

df[df['Salary'].between(60000, 70000)]
```

```python
#duplicate
df[df['First Name'].duplicated(keep='last')]
df[-df['First Name'].duplicated(keep=False)] # - is like not !
```

```python
df.drop_duplicates(subset=['First Name'], keep= False)
```
