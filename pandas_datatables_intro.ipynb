Ranks
In data analysis, the .rank() function is like a judge at a competition: it looks at a list of numbers and assigns each one a position (1st, 2nd, 3rd, etc.) based on its value relative to the others.

```python
nba = pd.read_csv('csv_files/nba.csv').dropna(how='all')
nba['Salary'] = nba['Salary'].fillna(0).astype('int')
nba.head()
```

```python
nba['Salary Rank'] = nba['Salary'].rank(ascending=False)
```

```python
nba.sort_values(by='Salary', ascending=False)
```
