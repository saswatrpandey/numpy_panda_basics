```python
import pandas as pd
```

```python
chicago = pd.read_csv('csv_files/chicago.csv', index_col='Name').dropna(how = 'all')
chicago.head()
```

```python
chicago.info()
```

```python
print(f"Unique Values: {chicago['Department'].nunique()}, Count: {chicago["Department"].count()}")
```

```python
chicago['Department'] = chicago['Department'].astype('category')
chicago.info()
```

String Method-> lower, upper, title, len

```python
'hello world'.title()
```

```python
chicago['Position Title'] = chicago['Position Title'].str.title()
chicago.head()
```

```python
chicago['Department'] = chicago['Department'].str.replace('MGMNT', 'MANAGEMENT')
```

```python
chicago['Employee Annual Salary'] = chicago['Employee Annual Salary'].str.replace('$', '').astype(float)
```

```python
chicago['Employee Annual Salary'].mean()
```

```python
mask = chicago['Position Title'].str.lower().str.contains('water') #startswith same functionality
mask2 = chicago['Position Title'].str.lower().str.startswith('water') #startswith same functionality
mask3 = chicago['Position Title'].str.lower().str.endswith('ist') #startswith same functionality


chicago[mask]
chicago[mask2]
chicago[mask3]
```

```python
chicago.index = chicago.index.str.strip().str.title()
chicago.head()
```

```python
chicago.columns = chicago.columns.str.upper()
chicago.head()
```
