```python
import sympy as sym
import numpy as np
from IPython.display import display, Math
sym.init_printing()
```

```python
x, y = sym.symbols('x, y')
display(x**y)
```

```python
# expression = symbolic, result = numeric
expression = sym.sqrt(2)
result = np.sqrt(2)

# Use sym.latex() to get the math symbol, and :.4f to trim the decimals
display(Math(f'{sym.latex(expression)} = {result:.4f}'))
```
