```python
import pandas as pd
```

Series


```python
places = ['Kathmandu', 'Lalitpur','Bhaktapur','Pokhara']

s = pd.Series(places)
print('values')
print(s.values)
print('keys')
print(s.index)
```

```python
#creating a series from dictionary
d_vals = {"hi":"Hello", 
          "ty": "Thank you",
          "btw": "By the way"}
pd.Series(d_vals)
```

Intro to Methods

```python
prices = [2.89, 45.66, 110.36]
s_new = pd.Series(prices)
print(s_new.sum())
print(s_new.product())
print(s_new.mean())
```

```python
fruits = ["Apple", "Banana", "Cherry", "Lemon"]
weekdays = ["Sunday", "Monday", "Tuesday", "Thursday"]

# pd.Series(fruits, weekdays)
pd.Series(data=fruits, index=weekdays)
```

Import Series with csv

```python
df_csv = pd.read_csv('csv_files/data.csv', usecols={"Duration"}) #prints out as dataframe
series_data = df_csv.squeeze("columns")
series_data.head(10) #tail same concept
#some functions
# len(series_data)
# sorted(series_data)
# dir(series_data)
# list(series_data)
# min(series_data)
# max(series_data)
```

```python
series_data.is_unique
series_data.ndim
series_data.shape
series_data.size
series_data.name = "Data" #modify name of the series
series_data
```

```python
#series_data.sort_values() #brand new series not just a view
series_data.sort_values(ascending=False).tail(10) 
```

```python
#traditional
# s_data_sorted = series_data.sort_values();
# s_data_sorted

series_data.sort_values(ascending=False, inplace=True) #inplace will actually change the main series now the series is sorted permanently
series_data
```

```python
series_data.sort_index(inplace=True) #This will sort the series in by index reversing the sort done above
series_data
```

```python
# 3 in [1,2,3,4]

60 in series_data #false because it checks the series' index rather than values
60 in series_data.values

series_data[[1,4,30]] #range of indexes value
series_data[-30:15]
```

```python
new_df_series = pd.read_csv('csv_files/data.csv',usecols={"Duration", "Pulse"}, index_col='Pulse') #override index with 'Pulse' value
new_series = new_df_series.squeeze("columns")
new_series.sort_index(inplace=True)
new_series
```

```python
new_series.get(key=102, default='No Data found')
```

Math Methods

```python
print(new_series.count()) #ignores null values
print(len(new_series)) #does not ignore null values
new_series.describe()

print('Index with max Value: ', new_series.idxmax())
print('Index with min Value: ', new_series.idxmin())
```

```python
new_series.value_counts() #counts repeated values in series
```

Apply method-> by calling function or lambda expression. 

```python
def classify_performance(number):
    if(number < 100):
        return 'Ok'
    else:
        return 'Normal'

new_series.apply(classify_performance)

new_series.apply(lambda pulse: pulse + 1)
```

**.map() Method**
Used when working with two or more datasets. It maps the values of 1 series to another series.


```python
s = pd.Series(['cat', 'cow', 'dog'])
print("Series:\n", s)
print("Mapping: ")
s.map({'cat': 'kitten', 'cow': 'calf'})
```
