```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
#Functional
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
plt.show() #optional prints anyways
```

```python
#plt.subplot(nrows, ncols, index)
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#Object oriented
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x,y)
axes.set_xlabel('X Level')
axes.set_xlabel('Y Level')
axes.set_title('Title')

```

```python
fig = plt.figure();

axes1 = fig.add_axes([0.1,0.1, 0.8,0.8])
axes2 = fig.add_axes([0.2,0.5, 0.4,0.3])

axes1.plot(x,y)
axes2.plot(y,x)

axes1.set_title('Larger')
axes2.set_title('Smaller')

```

```python
fig, axes = plt.subplots(nrows = 4, ncols = 3)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

for current_axis in axes:
    current_axis.plot(x,y)
```

```python
axes #an array value
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout() #Good thing to always put this.
```

Figure Size and DPI

```python
# 1. Use subplots() instead of figure() to get the axes array
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(6, 4)) 

# 2. Plotting on the first subplot (index 0)
axes[0].plot(x, y, label="x vs y")
axes[1].plot(y, x, label="y vs x") # This adds a second line to the SAME top plot

# Optional: Add labels or titles to make it readable
axes[0].set_title("Top Plot")
plt.tight_layout()
plt.show()
```

```python
fig.savefig('my_pic.png', dpi = 200) #save figures
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label='x square')
ax.plot(x, x**3, label='x cubed')

#loc will determine the placement of the legend
#ax.legend(loc = 10)
ax.legend(loc = [0.1,0.1])
```

Plot Styling

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# ax.plot(x, y, color= '#FF8C00')
#alpha is opacity
ax.plot(x, y, color= 'purple', 
        linewidth=0.5, alpha=0.5, ls='--', 
        marker='o', markersize=20, markerfacecolor='yellow', markeredgewidth=3, markeredgecolor='green') #ls-> linestyle

#set limit
ax.set_xlim([0,1])
ax.set_ylim([0,1])

```
